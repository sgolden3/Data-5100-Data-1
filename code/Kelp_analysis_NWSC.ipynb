## Add one and two years ago

```python
def get_historical_temperatures(current_date, df):
    """
    Finds historical temperatures for the same month and day in the previous year and two years prior.

    Args:
        current_date (pd.Timestamp): The date for which to find historical temperatures.
        df (pd.DataFrame): The DataFrame containing 'date' and 'temperature' columns.

    Returns:
        tuple: (temp_one_year_prior, temp_two_years_prior) or (NaN, NaN) if not found.
    """
    if pd.isna(current_date): # Handle NaT dates if any
        return pd.NA, pd.NA

    # Calculate target dates for one and two years prior
    one_year_ago = current_date - pd.DateOffset(years=1)
    two_years_ago = current_date - pd.DateOffset(years=2)

    # Helper to get temperature for a specific date, returning NaN if not found
    def get_temp_for_date(target_date, df_filtered):
        # Filter for the exact year, month, and day to ensure comparison is robust
        match = df_filtered[
            (df_filtered['date'].dt.year == target_date.year) 
            #(df_filtered['date'].dt.month == target_date.month)
        ]
        if not match.empty:
            return match['avg_temp'].iloc[0]
        return pd.NA

    # Get temperature for one year prior
    temp_one_year_prior = get_temp_for_date(one_year_ago, df)

    # Get temperature for two years prior
    temp_two_years_prior = get_temp_for_date(two_years_ago, df)

    return temp_one_year_prior, temp_two_years_prior

print("Function 'get_historical_temperatures' defined successfully.")
```

```python
polnell[['temperature_1_year_prior', 'temperature_2_years_prior']] = tester.apply(lambda row: get_historical_temperatures(row['date'], tester), axis=1, result_type='expand')
poss_pt[['temperature_1_year_prior', 'temperature_2_years_prior']] = poss_pt.apply(lambda row: get_historical_temperatures(row['date'], poss_pt), axis=1, result_type='expand')

nobo[['temperature_1_year_prior', 'temperature_2_years_prior']] = nobo.apply(lambda row: get_historical_temperatures(row['date'], nobo), axis=1, result_type='expand')
```

```python
time_cols = ['temperature_1_year_prior','temperature_2_years_prior']
for col in time_cols:
    polnell[col] = pd.to_numeric(polnell[col], errors='coerce')

time_cols = ['temperature_1_year_prior','temperature_2_years_prior']
for col in time_cols:
    poss_pt[col] = pd.to_numeric(poss_pt[col], errors='coerce')

time_cols = ['temperature_1_year_prior','temperature_2_years_prior']
for col in time_cols:
    nobo[col] = pd.to_numeric(nobo[col], errors='coerce')
```
